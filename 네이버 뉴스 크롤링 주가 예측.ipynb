```python
# 제목 수준
import requests
from bs4 import BeautifulSoup

keyword = input("검색어 입력: ")

headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
# response = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query="+keyword,headers = headers)
response = f'https://search.naver.com/search.naver?sm=tab_hty.top&where=news&query={keyword}&pd=12'
html = response.text

# 시간 설정을 url 선에서 처리 하거나
# 일단 불러와 놓고 시간도 크롤링해서 처리하거나...
# 'pd = 12'가 최근 6시간 의미
# 그래서 관련도 순 6시간 이내

soup = BeautifulSoup(html,'html.parser')
# html에서 뉴스 제목 부분의 클래스 명
# a 태그의 정보가 리스트 형태로 담긴다!

links = soup.select("div.group_news > ul.list_news > li div.news_area > div.news_info > div.info_group > a.info")
# '.info' : info라는 클래스 명으로 찾는다.
# 'tag' : 태그이름으로 찾음
# '#id' : 아이디 이름을 찾음
# '상위태그이름>자식태그>자식태그' : 부모 자식간의 태그 조회
# https://wonhwa.tistory.com/52 참고

urls = []


for link in links:
    title = link.text # 태그 안에 텍스트 요소를 가져옴
    url = link.attrs['href']    # href의 속성값을 가져옴
    urls.append(url)
#     soup3 = BeautifulSoup(requests.get(url,headers=headers).text,'html.parser')
#     #for j in soup3.find_all('div',{'class':'viewer'}):
#     for j in soup3.select(".viewer"):
#         print(j.text)

naver_urls = []
naver_articles = []
    
for url in urls:
    if "news.naver.com" in url:
        naver_urls.append(url)
    else:
        pass

for url in naver_urls:
    response = requests.get(url,headers = headers)
    html = response.text
    naver_soup = BeautifulSoup(html,'html.parser')
    for j in naver_soup.select('#dic_area'):
        naver_articles.append(j.text)
        print(j.text)

print(naver_urls)
```

```python
from konlpy.tag import Kkma, Hannanum
from konlpy.utils import pprint

kkma = Kkma()
pprint(kkma.nouns(naver_articles[0]))
```

```python
hannanum=Hannanum()
print(hannanum.nouns(naver_articles[0]))
```

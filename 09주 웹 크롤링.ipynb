```python
#urllib 활용
from urllib.request import urlopen
import sys
crawling_url = "https://hanbit.co.kr/store/books/full_book_list.html"
response = urlopen(crawling_url)
encoding = response.info().get_content_charset(failobj="utf-8")
text=response.read().decode(encoding)
print("encoding:",encoding, file=sys.stderr)
print(text)
```

```python
# 정규 표현식
import re

str = "우리가 만약 Tom과 만나게 되면 jane은 뭐라고 할까? 윌리엄은? 그러면 picnic은? 그럼 K는?"
re.findall('[A-Z][a-z]+',str) # 대문자 하나, 소문자 하나 이상
```

```python
str = "니가 185번 버스를 타고 가다가 130번으로 갈아타면 될 거 같은데, 비용은 1,500원이 들거야"
print(re.findall('[\d]+',str))
# ,로 금액을 분할해버리네? 포함 시켜보자!
print(re.findall('[\d]+,[\d]+|[\d]+',str))
```

```python
str2 = "우리가 만약 Tom과 만나게 되면 Jane은 뭐라고 할까? 윌리엄은? 그러면 Edward은? 그럼 Kane는?"

re_iter=re.finditer('[A-Z][a-z]+',str2)
for s in re_iter:
    print(s)
```

```python
str1 = "강"
str2 = "우리나라의 4대 방어선은 압록강, 청천강, 두만강, 임진강이다"
s=re.split(str1,str2)
for t in s:
    tt=re.split(' ',t)
    if(len(tt)>1):
        print(tt[-1]+"강")
```

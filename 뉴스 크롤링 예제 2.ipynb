```python
import os, re
import requests

import urllib.request as ur
from bs4 import BeautifulSoup as bs
```

```python
# 기사 하나
article1 = 'https://n.news.naver.com/mnews/article/421/0006778347?sid=101'

soup = bs(ur.urlopen(article1).read(),'html.parser')
soup
```

```python
contents=[]
for i in soup.find_all('div',{"class":'go_trans _article_content'}):
    contents.append(i.text)
    
```

```python
contents
```

```python
title = soup.find('div',{"class":'media_end_head_title'})
print(title.text)
```

```python
# 여러 기사 실습
url = "https://search.naver.com/search.naver?where=news&query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&sm=tab_tmr&nso=so:r,p:all,a:all&sort=0"
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
soup2 = bs(ur.urlopen(url).read(),'html.parser')
soup2
```

```python
headlines = soup2.find_all('a',{'class':'news_tit'})
titles2 = []
for i in headlines:
    titles2.append(i.text)
```

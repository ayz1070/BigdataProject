```python
import os, re
import requests

import urllib.request as ur
from bs4 import BeautifulSoup as bs
```

```python
os.chdir(r'C:\Users\ahn\Bigdata')
news = 'https://news.daum.net/'

soup = bs(ur.urlopen(news).read(),'html.parser')
```

```python
soup
```

```python
soup.find_all('div',{"class":"item_issue"})
```

```python
texts = []
for i in soup.find_all('div',{"class":"item_issue"}):
    texts.append(i.text)
```

```python
texts
```

```python
# 포털 사이트의 메인화면은 수많은 링크의 집합
# <a> 태그 : <a href = "링크할 URL 주소">하이퍼링크 텍스트</a>
# 새로운 URL 주소를 자동으로 연결하는 것을 가능하게 해준다.
soup.find_all('a')
```

```python
links = []
for i in soup.find_all('a'):
    links.append(i.get('href'))
links
```

```python
arr = []
for i in soup.find_all('div',{'class':'item_issue'}):
    arr.append(i.find_all('a')[0])
arr
```

```python
arr2 = []
for i in soup.find_all('div',{"class":"item_issue"}):
    arr.append(i.find_all('a')[0].get('href'))
arr2
```

```python
# columns 와 index 값을 줘보자!

# (2) 2D ndarray로부터 생성 
# index(행) : date 정보 : 2020녀 1월 1일 ~ 2020년 1월 10일
# column (열) : A,B,C

# 2020년 부터, 10개, 주기는 Date (M : 달 , Y : 년)
index = pd.date_range('2020-01-01',periods=10,freq='D')
df = pd.DataFrame(
    np.random.rand(10,3), index, columns = ["A","B","C"]
)

df
```
